```python
import pandas as pd
import os



# Step 2: Define the path to the folder containing the Excel files
input_folder_path = '/content/drive/MyDrive/GCNN SECOND TRIAL!!/deoxy data/Neutral Deoxy'
output_folder_path = '/content/drive/MyDrive/GCNN SECOND TRIAL!!/Neutral Deoxy Norm'

# Step 3: Create the output folder if it doesn't exist
if not os.path.exists(output_folder_path):
    os.makedirs(output_folder_path)

# Step 4: List all Excel files in the input folder
excel_files = [f for f in os.listdir(input_folder_path) if f.endswith('.xlsx')]

# Step 5: Function to perform max-min normalization
def max_min_normalization(df):
    return (df - df.min()) / (df.max() - df.min())

# Step 6: Process each Excel file
for file in excel_files:
    input_file_path = os.path.join(input_folder_path, file)
    # Read the Excel file
    df = pd.read_excel(input_file_path,header=None)

    # Normalize the data
    normalized_df = max_min_normalization(df)

    # Save the normalized data to a new Excel file in the output folder
    output_file_path = os.path.join(output_folder_path, f'normalized_{file}')
    normalized_df.to_excel(output_file_path, index=False,header=False)

    print(f'Normalized {file} and saved to {output_file_path}')

print("Normalization of all files completed.")
```

```python
#COUNT NUMBER OF FILES IN A DRIVE FOLDER

import os

def count_files_in_folder(folder_path):
    try:
        # List all entries in the given folder
        entries = os.listdir(folder_path)

        # Filter out directories, keeping only files
        files = [entry for entry in entries if os.path.isfile(os.path.join(folder_path, entry))]

        # Return the count of files
        return len(files)

    except FileNotFoundError:
        return "The specified folder was not found."
    except PermissionError:
        return "Permission denied to access the specified folder."
    except Exception as e:
        return f"An error occurred: {e}"

# Example usage
folder_path = "/content/drive/MyDrive/GCNN SECOND TRIAL!!/Neutral Oxy Norm"
print(f"Number of files in '{folder_path}': {count_files_in_folder(folder_path)}")
```

```python
#File Deletion
import os
import glob

def delete_neutral_files(directory):
    # Construct the pattern for the files to be deleted
    pattern = os.path.join(directory, '*neutral*')

    # Get the list of all files matching the pattern
    files_to_delete = glob.glob(pattern)

    # Iterate through the list and delete each file
    for file_path in files_to_delete:
        try:
            os.remove(file_path)
            print(f"Deleted: {file_path}")
        except Exception as e:
            print(f"Error deleting {file_path}: {e}")

# Example usage
directory_path = '/content/drive/MyDrive/GCNN SECOND TRIAL!!/oxygenated data/Neutral Oxy'
delete_neutral_files(directory_path)
```
